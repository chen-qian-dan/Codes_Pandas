# Visualization in Pandas

```python
# Series and DataFrame line chart
import numpy as np
import pandas as pd
# let Jupyter notebook know that we put the chart on here.
%matplotlib inline   

ts = pd.Series(np.random.randn(50), index = pd.date_range('today', periods = 50))

ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(50, 4), index = ts.index, columns = ['A', 'B', 'X', 'Y'])
df = df.cumsum()
df.plot()
```

# Practice example
# Remove repeated data using pandas

```python
df = pd.DataFrame({'A': [1, 2, 2, 2, 4, 4, 5, 5, 6, 6, 7, 8, 8]})
print(df)
```

```python
df = pd.DataFrame({'A': [1, 2, 2, 2, 4, 4, 5, 5, 6, 6, 7, 8, 8]})
print(df)
print('------------------')
# the default parameter of shift is 1; 
df2 = df.loc[df['A'].shift() != df['A']]
print(df2)
```

```python
df = pd.DataFrame({'A': [1, 2, 2, 2, 4, 4, 5, 5, 6, 6, 7, 8, 8]})
print(df)
print('------------------')
df2 = df.loc[df['A'].shift(2) != df['A']]
print(df2)
```

```python
print(df)
print('------------------')
df2 = df.loc[df['A'].shift(-1) != df['A']]
print(df2)
```
